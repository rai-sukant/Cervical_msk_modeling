```python
import pandas as pd
import numpy as np

# File paths
input_path = r"C:\Users\ASUS\Desktop\Trial_01_angle.csv"
output_path = r"C:\Users\ASUS\Desktop\Trial_01_angle_100Hz_rad.csv"

# Read CSV
df = pd.read_csv(input_path)

# -----------------------------
# 1. Fix time column to 100 Hz
# -----------------------------
fs = 100
dt = 1 / fs
df.iloc[:, 0] = np.arange(0, len(df) * dt, dt)

# ----------------------------------
# 2. Convert angle columns to radians
# ----------------------------------
df.iloc[:, 1:] = np.deg2rad(df.iloc[:, 1:])

# -----------------------------
# 3. Save updated CSV
# -----------------------------
df.to_csv(output_path, index=False)

print("Time updated to 100 Hz and angles converted to radians.")
```
